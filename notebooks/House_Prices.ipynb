```python
# import io
import pandas as pd

# path = "\home\mlops-alura\datasets\casas.csv"
path = "../datasets/casas.csv"
df = pd.read_csv(path,sep=',')
```

```python
df.head()
```

```python
X = df.drop('preco', axis=1)
y = df['preco']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
dados_input = [120,2001,2]
```

```python
modelo.predict([dados_input])
```

```python
import pickle

# salvar o modelo
with open('../models/modelo.sav', 'wb') as file:
    pickle.dump(modelo, file)
```
